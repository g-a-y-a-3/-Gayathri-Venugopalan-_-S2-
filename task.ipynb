```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('/content/car_price_prediction.csv')
print(dataset.head())
print(dataset.describe())
```

```python
import matplotlib.pyplot as plt
plt.hist(dataset['Prod. year'])
plt.xlabel('Prod. year')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
plt.bar(dataset['Fuel type'], dataset['Price'])
plt.xlabel('Fuel type')
plt.ylabel('Price')
plt.title('Bar Plot')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
plt.boxplot(dataset['Cylinders'])
plt.xlabel('Cylinders')
plt.title('Box Plot of Cylinders')
plt.grid(True)
plt.show()
plt.scatter(dataset['Cylinders'], dataset['Airbags'])
plt.xlabel('Cylinders')
plt.ylabel('Airbags')
plt.title('Scatter Plot of Cylinders vs. Airbags')
plt.show()
```

```python
print(dataset.isnull().sum())
df = dataset.dropna()
print(df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns_to_standardize = ['Airbags', 'Price', 'Cylinders']
dataset[columns_to_standardize] = scaler.fit_transform(dataset[columns_to_standardize])
print(dataset[columns_to_standardize])
```

```python
correlation_matrix = dataset.corr()
print(correlation_matrix)
```
